```julia
using LinearAlgebra, Plots, Random, Statistics, Plots.Measures
using LaTeXStrings
```

```julia
t = 0:0.01:4*π
sigma = 0.5
y = cos.(t) + sigma * randn(length(t))

FontSize = 15
plot_font = "Computer Modern"
default(fontfamily=plot_font,
        linewidth=2, framestyle=:box, label=nothing, grid=false)
p_cosine = plot(size=(500,300),y,ylabel=L"\textrm{Signal}",xlabel=L"\textrm{Time}",label="",title="Noisy Cosine Wave", 
    xtickfontsize=FontSize,ytickfontsize=FontSize,xguidefontsize=FontSize,yguidefontsize=FontSize,legendfontsize=FontSize,
titlefontsize = FontSize, bottom_margin = 5mm,
)
savefig(p_cosine, "/Users/kgilman/Desktop/noisy_cosine.pdf")
plot!(p_cosine)
```

```julia
t = 0:0.01:4*π
sigma = 0.5
y = sin.(t) + sigma * randn(length(t))

FontSize = 15

p_sine = plot(size=(500,300),y,ylabel="Signal",xlabel="Time",label="",title="Noisy Sine Wave", 
    xtickfontsize=FontSize,ytickfontsize=FontSize,xguidefontsize=FontSize,yguidefontsize=FontSize,legendfontsize=FontSize,
titlefontsize = FontSize, bottom_margin = 5mm)
savefig(p_sine, "/Users/kgilman/Desktop/noisy_sine.pdf")
plot!(p_sine)
```

```julia
t = -2:0.01:2
sigma = 0.5
y = exp.(t) + sigma * randn(length(t))

FontSize = 15

p_exp = plot(size=(500,300),y,ylabel="Signal",xlabel="Time",label="",title="Noisy Exponential", 
    xtickfontsize=FontSize,ytickfontsize=FontSize,xguidefontsize=FontSize,yguidefontsize=FontSize,legendfontsize=FontSize,
titlefontsize = FontSize, bottom_margin = 5mm)
savefig(p_exp, "/Users/kgilman/Desktop/noisy_exp.pdf")
plot!(p_exp)
```

```julia
square(x) = if(abs(x) <=1)  1 else 0 end
```

```julia
t = -2:0.01:2
sigma = 0.1
y = square.(t) + sigma * randn(length(t))

FontSize = 15

p_impulse = plot(size=(500,300),y,ylabel="Signal",xlabel="Time",label="",title="Noisy Impulse", 
    xtickfontsize=FontSize,ytickfontsize=FontSize,xguidefontsize=FontSize,yguidefontsize=FontSize,legendfontsize=FontSize,
titlefontsize = FontSize, bottom_margin = 5mm)
savefig(p_impulse, "/Users/kgilman/Desktop/noisy_impulse.pdf")
plot!(p_impulse)
```
